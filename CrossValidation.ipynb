```python
# Import libraries
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load dataset
digits = load_digits()

X = digits.data
y = digits.target
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
#Logistic Regression

lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)

print("Logistic Regression Test Score:", lr.score(X_test, y_test))
```

```python
#Support Vector Machine

svc = SVC()
svc.fit(X_train, y_train)

print("SVC Test Score:", svc.score(X_test, y_test))
```

```python
# Random Forest

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)

print("Random Forest Test Score:", rf.score(X_test, y_test))
```

```python
# Cross Validation

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

print("\nCross Validation Scores:")

print("Logistic Regression CV:",
      cross_val_score(LogisticRegression(max_iter=5000), X, y, cv=kfold).mean())

print("SVC CV:",
      cross_val_score(SVC(), X, y, cv=kfold).mean())

print("Random Forest CV:",
      cross_val_score(RandomForestClassifier(n_estimators=100), X, y, cv=kfold).mean())
```
