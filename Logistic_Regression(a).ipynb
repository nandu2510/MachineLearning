**It is used to solve the classification problems**




*  Binary classification->only have 2 options to choose answer like yes or no
*  Multiclass classification->This has more than 2 values as the options to choose as an answer



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "age": [22,25,28,30,32,35,38,40,42,45,48,50,55,58,60],
    "bought_insurance": [0,0,1,0,1,1,0,1,1,0,1,1,1,1,1]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python

df.head()
```

```python
df.shape
```

Visualizing the data

```python
plt.scatter(x=df["age"],y=df["bought_insurance"],marker="+",color="red")
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.show()
```

Splitting data into input and output

```python
X=df[['age']]
Y=df['bought_insurance']
```

Training data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1,random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

**Create Logistic Regression model**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

**Train the model**

```python
model.fit(X_train, y_train)
```

**Predict**

```python

predictions = model.predict(X_test)
print(predictions)
```

**Accuracy**

```python
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```
