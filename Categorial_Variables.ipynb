**The below is using pandas**

```python
import pandas as pd
data = {
    'town': [
        'Monroe Township', 'Robbinsville', 'West Windsor', 'Monroe Township',
        'Robbinsville', 'West Windsor', 'Monroe Township', 'Robbinsville',
        'West Windsor', 'Monroe Township', 'Robbinsville', 'West Windsor'
    ],
    'area': [2100, 1600, 2400, 1400, 3000, 1980, 2200, 1850, 2100, 1700, 2000, 2300],
    'price': [500000, 350000, 600000, 300000, 650000, 450000, 520000, 400000, 480000, 360000, 420000, 550000]
}
df = pd.DataFrame(data)
print(df)
```

```python
dumies=pd.get_dummies(df.town)
dumies
```

```python
merged=pd.concat([df,dumies],axis=1)
merged
```

```python
mergelast = merged.drop(['town'], axis=1)
print(mergelast)
```

**The below is using sklearn**

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'town': ['Monroe Township', 'Robbinsville', 'West Windsor', 'Monroe Township',
             'Robbinsville', 'West Windsor', 'Monroe Township', 'Robbinsville',
             'West Windsor', 'Monroe Township', 'Robbinsville', 'West Windsor']
})
X = df[['town']].values  # 2D array
ohe = OneHotEncoder(drop='first', sparse_output=False)  # use sparse_output=False to get numpy array directly
X_encoded = ohe.fit_transform(X)
print(X_encoded)
```
