```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {
    "Area": [1200, 1500, 1800, 2000, 2500],
    "Bedrooms": [2, 3, np.nan, 4, 4],  # np.nan for missing value
    "Age": [10, 8, 5, 3, 1],
    "Price": [45, 65, 80, 100, 130]
}
```

```python
df = pd.DataFrame(data)
df.to_csv("house_multiple.csv", index=False)
df
```

```python
import math
median_bedrooms = math.floor(df.Bedrooms.median())
median_bedrooms
```

```python
df.Bedrooms.fillna(median_bedrooms)
```

```python
df['Bedrooms'] = df['Bedrooms'].fillna(median_bedrooms)
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area','Bedrooms','Age']],df.Price)
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
