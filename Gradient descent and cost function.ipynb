```python
import numpy as np
names = np.array(["A", "B", "C", "D"])
maths = np.array([60, 70, 80, 90]).reshape(-1,1)
cs = np.array([65, 75, 85, 95])
```

```python
maths = maths / 100
cs = cs / 100
```

```python
m=0
b=0
learning_rate = 0.3
iterations =10000

n=len(maths)
```

```python
for i in range(iterations):
   y_predictes=m*maths +b  #predicting the cs marks
   cost=(1/n)*sum((y_predictes)**2) #cost function(mean sqaured error)
 # gradients
   md = -(2/n) * sum(maths * (cs - y_predictes))
   bd = -(2/n) * sum(cs - y_predictes)
```

```python
from re import L
b=b-learning_rate*md
m=m-learning_rate*bd
```

```python
print("Final slope (m):", m)
print("Final bias (b):", b)
```

```python
print("Predicted cs marks:")
for i in range(n):
  print("Name:",names[i],  "| Maths:", maths[i],
        "| Actual CS:", cs[i],
        "| Predicted CS:", m * maths[i] + b)
```
